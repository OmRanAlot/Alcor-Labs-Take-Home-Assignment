```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = Path("../data/analysis/diff_ssim.csv")
df = pd.read_csv(csv_path)
df.head()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df["timestamp_sec"], df["ssim"], marker="o", linestyle="-")
plt.plot(df["timestamp_sec"], df["diff"]/100, marker="o", linestyle="-")
plt.title("Difference in SSIM Over Frames")
plt.xlabel("Frame")
plt.ylabel("Value")
plt.legend(["SSIM", "Difference"])
plt.grid()
plt.show()
```
